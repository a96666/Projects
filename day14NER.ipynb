```python

!pip install spacy
!python -m spacy download en_core_web_sm

import spacy
import pandas as pd
from spacy import displacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Match report text
text = """
Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London.
Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024.
"""

# Process text with spaCy
doc = nlp(text)

print("===== Named Entity Recognition Results =====\n")

# Store entities
entities = []

for ent in doc.ents:
    print(f"{ent.text} --> {ent.label_}")
    entities.append((ent.text, ent.label_))

# Convert to structured DataFrame
df = pd.DataFrame(entities, columns=["Entity", "Type"])

print("\n===== Structured Entity Table =====")
print(df)

# Categorize entities for insights
teams = [ent.text for ent in doc.ents if ent.label_ == "ORG"]
players = [ent.text for ent in doc.ents if ent.label_ == "PERSON"]
locations = [ent.text for ent in doc.ents if ent.label_ in ["GPE","LOC","FAC"]]
dates = [ent.text for ent in doc.ents if ent.label_ == "DATE"]
numbers = [ent.text for ent in doc.ents if ent.label_ == "CARDINAL"]

print("\n===== Extracted Insights =====")
print("Teams:", teams)
print("Players:", players)
print("Locations:", locations)
print("Dates:", dates)
print("Numbers:", numbers)

# Visualize entities (works in Jupyter/Colab)
displacy.render(doc, style="ent", jupyter=True)
```
