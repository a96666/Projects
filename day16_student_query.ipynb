```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Student query
query = "beginner-friendly Python tutorials"

# Course catalog
courses = [
    "Introduction to Programming with Python",
    "Advanced Data Structures in C++",
    "Machine Learning with Python",
    "Web Development with HTML and CSS"
]

# Example embeddings (normally generated by an embedding model)
query_embedding = np.array([[0.9, 0.8, 0.7]])

course_embeddings = np.array([
    [0.85, 0.75, 0.65],  # Python intro
    [0.2, 0.3, 0.1],     # C++ course
    [0.7, 0.6, 0.8],     # ML Python
    [0.1, 0.2, 0.3]      # Web course
])

# Calculate cosine similarity
similarities = cosine_similarity(query_embedding, course_embeddings)

# Get the most relevant course
best_match_index = np.argmax(similarities)

print("Student Query:", query)
print("Recommended Course:", courses[best_match_index])
print("Similarity Score:", similarities[0][best_match_index])
```
