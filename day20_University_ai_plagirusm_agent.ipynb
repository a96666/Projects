```python
# Step 1: Tools

class PlagiarismTools:

    def flag_plagiarism(self):
        return "⚠️ Potential Plagiarism Detected"

    def mark_original(self):
        return "✅ Original Work"
```

```python
# Step 2: LLM Brain

class PlagiarismLLM:

    def think(self, essay):
        print("🧠 Thinking...")

        if "copy" in essay.lower() or "verbatim" in essay.lower():
            return "Essay contains suspicious copied content"
        else:
            return "Essay seems original"

    def plan(self, thought):
        print("📌 Planning...")

        if "suspicious" in thought:
            return "flag"
        else:
            return "original"

    def answer(self, result):
        return f"📢 Decision: {result}"
```

```python
# Step 3: ReAct Agent

def plagiarism_agent(essay):

    llm = PlagiarismLLM()
    tools = PlagiarismTools()

    # Step 1: Think
    thought = llm.think(essay)
    print("💭 Thought:", thought)

    # Step 2: Plan
    action = llm.plan(thought)
    print("⚙️ Action Chosen:", action)

    # Step 3: Act
    if action == "flag":
        result = tools.flag_plagiarism()
    else:
        result = tools.mark_original()

    # Step 4: Final Answer
    final = llm.answer(result)
    print(final)
```

```python
# Example 1 (Plagiarized)
essay1 = "This essay contains verbatim copied content from internet"
plagiarism_agent(essay1)

print("\n------------------\n")

# Example 2 (Original)
essay2 = "This essay is written in my own words with unique ideas"
plagiarism_agent(essay2)
```
