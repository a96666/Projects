```python
import numpy as np

# Patient symptom vector
patient = np.array([0.2, 0.8, -0.3, 0.5])

# Condition symptom profile vector
condition = np.array([0.1, 0.9, -0.2, 0.4])

# Step 1: Dot product
dot_product = np.dot(patient, condition)

# Step 2: Magnitude of vectors
patient_mag = np.linalg.norm(patient)
condition_mag = np.linalg.norm(condition)

# Step 3: Cosine similarity
cosine_similarity = dot_product / (patient_mag * condition_mag)

print("Patient Vector:", patient)
print("Condition Vector:", condition)
print("Dot Product:", dot_product)
print("Patient Magnitude:", patient_mag)
print("Condition Magnitude:", condition_mag)
print("Cosine Similarity:", cosine_similarity)
```
