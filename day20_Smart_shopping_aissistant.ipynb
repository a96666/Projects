```python
# Step 1: Tools

def search_cakes():
    # Dummy dataset (like your notebook style)
    return [
        {"name": "Chocolate Truffle Cake", "price": 450, "rating": 4.5},
        {"name": "Dark Chocolate Cake", "price": 550, "rating": 4.7},
        {"name": "Chocolate Fudge Cake", "price": 400, "rating": 4.2},
        {"name": "Premium Belgian Cake", "price": 700, "rating": 4.9}
    ]

def filter_by_budget(cakes, budget):
    return [cake for cake in cakes if cake["price"] <= budget]

def choose_best_cake(cakes):
    # choose highest rating
    if not cakes:
        return "No cake found under budget "
    best = max(cakes, key=lambda x: x["rating"])
    return best
```

```python
# Step 2: Agent Brain

def agent_brain(user_query):
    print(" Thinking...")

    if "cake" in user_query.lower():
        return ["search_cakes", "filter_budget", "choose_best"]
    else:
        return ["unknown"]
```

```python
# Step 3: Execute Plan

def execute_plan(plan, budget):
    data = None

    for step in plan:
        if step == "search_cakes":
            print("🔍 Searching cakes...")
            data = search_cakes()

        elif step == "filter_budget":
            print(f"💰 Filtering cakes under ₹{budget}...")
            data = filter_by_budget(data, budget)

        elif step == "choose_best":
            print("🏆 Choosing best cake...")
            data = choose_best_cake(data)

    return data
```

```python
# Step 4: Final Agent

def shopping_agent(user_query, budget):
    print(f"📝 User Query: {user_query}")

    plan = agent_brain(user_query)
    print("📌 Plan:", plan)

    result = execute_plan(plan, budget)

    print("\n✅ Final Answer:")
    print(result)
```

```python
# Run example

shopping_agent("Find me the best chocolate cake", 500)
```
