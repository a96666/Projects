```python
"""📘 Scenario: Sports Event Analysis
Context:
You are part of a sports analytics team working for a global media company. Your manager has asked you to analyze match reports and press releases to identify key entities such as teams, players, locations, dates, and numbers. This information will help the company build automated summaries and highlight reels for fans.
Task:
You receive the following match report excerpt:
"Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London. Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024."

Your job is to run Named Entity Recognition (NER) using spaCy to automatically extract structured insights."""
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Match report text
text = """Manchester United defeated Real Madrid 3-2 at Wembley Stadium, London.
Cristiano Ronaldo scored twice, while Marcus Rashford netted the winning goal in July 2024."""

# Process the text
doc = nlp(text)

# Print detected entities
print("Named Entities:\n")

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# STEP 1: Import required library
import spacy

# STEP 2: Load spaCy model
nlp = spacy.load("en_core_web_sm")

# STEP 3: Define University Knowledge Base
knowledge_base = """
University Helpdesk Information

Hostel Admission Requirements:
- Admission confirmation letter
- Government ID proof
- Two passport size photographs

Course Registration:
- Registration happens online through the university portal
- Students must register before the semester deadline

Fee Payment:
- Fees can be paid through the student dashboard
- Online payment methods include debit card, credit card, and net banking

Examination Schedule:
- Semester exams usually start in December and May

Library Access:
- Students must carry their university ID card
"""

# STEP 4: Process knowledge base using spaCy
kb_doc = nlp(knowledge_base)

# STEP 5: Take user question
query = input("Ask a university helpdesk question: ")

# STEP 6: Process user query
query_doc = nlp(query)

# STEP 7: Calculate similarity between query and knowledge base
similarity = query_doc.similarity(kb_doc)

print("\nSimilarity Score:", similarity)

# STEP 8: Simple keyword-based answer retrieval
query_lower = query.lower()

if "hostel" in query_lower:
    print("\nHostel Admission Requirements:")
    print("- Admission confirmation letter")
    print("- Government ID proof")
    print("- Two passport size photographs")

elif "course" in query_lower or "registration" in query_lower:
    print("\nCourse Registration:")
    print("Registration happens online through the university portal before the semester deadline.")

elif "fee" in query_lower or "payment" in query_lower:
    print("\nFee Payment:")
    print("Fees can be paid through the student dashboard using debit card, credit card, or net banking.")

elif "exam" in query_lower:
    print("\nExamination Schedule:")
    print("Semester exams usually start in December and May.")

elif "library" in query_lower:
    print("\nLibrary Access:")
    print("Students must carry their university ID card.")

else:
    print("\nSorry, I could not find information related to your query.")
```
