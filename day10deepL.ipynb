```python
import numpy as np

# Input data (number of items)
X = np.array([1, 2, 3, 4, 5])

# Target output
y = np.array([0, 0, 1, 1, 1])

# Initialize parameters
weight = 0
bias = 0
learning_rate = 0.1
epochs = 10

# Step activation function
def step_function(z):
    if z >= 0:
        return 1
    else:
        return 0

# Training the perceptron
for epoch in range(epochs):
    print("Epoch:", epoch+1)

    for i in range(len(X)):

        # Linear equation
        z = weight * X[i] + bias

        # Prediction
        y_pred = step_function(z)

        # Error
        error = y[i] - y_pred

        # Update weight and bias
        weight = weight + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print("Input:", X[i],
              "Predicted:", y_pred,
              "Actual:", y[i],
              "Weight:", round(weight,2),
              "Bias:", round(bias,2))

    print()
```
